```python
Q1 = df['Age'].quantile(0.25) #Calculate the first quartile (25th percentile) of the Age column
Q3 = df['Age'].quantile(0.75) #Calculate the third quartile (75th percentile) of the Age column
IQR = Q3 = Q1 #Compute the Interquartile Range (IQR), used to detect outliers
lower_bound = Q1 - 1.5 * IQR #Calculate the lower limit for detecting outliers
upper_bound = Q3 + 1.5 * IQR #Calculate the upper limit for detecting outliers

df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)] #Filter the dataframe to remove Age values outside the IQR bounds
```

```python
Q1 = df['Fare'].quantile(0.25) #Calculate the first quartile (25th percentile) of the Fare column
Q3 = df['Fare'].quantile(0.75) #Calculate the third quartile (75th percentile) of the Fare column
IQR = Q3 - Q1 #Compute the Interquartile Range (IQR), used to detect outliers
lower_bound = Q1 - 1.5 * IQR #Calculate the lower limit for detecting Fare outliers
upper_bound = Q3 + 1.5 * IQR #Calculate the upper limit for detecting Fare outliers

df = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]  # Remove Fare values outside the IQR range from the dataframe
```

```python
from sklearn.preprocessing import MinMaxScaler  #Import MinMaxScaler to normalize numerical data to a range (0–1)
scaler = MinMaxScaler() #Create a MinMaxScaler object for scaling features
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']]) #Scale the Age and Fare columns so their values range between 0 and 1

df.head() #Display the first 5 rows of the dataframe to check the scaled values
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked', 'Pclass'], drop_first=True) #Convert categorical variables into numerical dummy/indicator variables

df.head() #Show the first 5 rows to verify the encoded categorical columns
```

```python
df['family_size'] = df['SibSp'] + df['Parch'] + 1 #Create a new feature 'family_size' by adding siblings/spouses, parents/children, plus the passenger

df[['SibSp', 'Parch', 'family_size']].head() #Display the related columns to verify the new family_size feature
```

```python
from sklearn.decomposition import PCA #Import Principal Component Analysis for dimensionality reduction

features = df.select_dtypes(include=['number']) #Select only numerical columns from the dataframe for PCA

pca = PCA(n_components=3) #Initialize PCA to reduce the dataset into 3 principal components
principal_components = pca.fit_transform(features) #Fit PCA on the numeric data and transform it into principal component values

df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3']) #Create a new dataframe containing the 3 principal components

df_pca.head() #Display the first 5 rows of the PCA-transformed dataset
```
